```python
dataset = pd.read_csv('https://raw.githubusercontent.com/fspot/INFMDI-721/master/lesson5/products.csv',delimiter =';')
dataset['currency'] = dataset.price.str.split(' ').str[1]
dataset['price'] = dataset.price.str.split(' ').str[0]
dataset.head()
```

```python
def get_currency(IP):
    response = requests.get('http://www.geoplugin.net/json.gp?ip='+IP)
    data = json.loads(response.text)
    if len(data) == 0:
        return 'NA'
    else:
        currency = data['geoplugin_currencyCode']
        return currency  
```

```python
for index in range(0,100):
        try:
            dataset.loc[index,'currency'] = get_currency(dataset.loc[index,'ip_address'])
        except:
            dataset.loc[index,'currency'] = 'NA'
dataset.head()
```
