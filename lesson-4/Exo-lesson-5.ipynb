```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import json
from datetime import datetime
from datetime import timedelta
```

```python
df = pd.read_csv('/home/rakotoniaina/TELECOM PARIS/INFMDI-721/lesson4/people.csv',delimiter =',')
df['last_seen'] = pd.to_datetime(df.last_seen, unit='s')
df.head()
```

```python
past = datetime.now() - timedelta(days=365)
def duree(x):
    if x < past:
        return True
    else:
        return False
    

df['inactive'] = df['last_seen'].apply(duree)

df.head()
```
