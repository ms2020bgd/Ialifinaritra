Faire un programme qui:

## 1. Liste des 10 villes les plus peuplées

```python
import requests
from bs4 import BeautifulSoup

url = "https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peupl%C3%A9es"
```

```python
def get_soup_from_url(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    return soup
```

```python
soup = get_soup_from_url(url)
table=soup.find('table').find_all('tr')[1::1]

city = []
ville =[]

for items in table:
    data = items.find_all(['th','td'])
    name = data[1].a.text
    city.append(name)

for i in range(10):
    ville.append(city[i])
    print(city[i])
```

## 2. Distance entre les villes ( en utilisant l'API )

```python
def get_distance_between(a , b):
    
    url_distance = f"https://fr.distance24.org/route.json?stops={a}|{b}"
    data = requests.get(url_distance).json()
    return data['distance']
```

```python
import itertools
import pandas as pd
def get_cities_combination(cities):
    product = list(itertools.product(cities, repeat=2))
    df = pd.DataFrame(product, columns=['origin', 'dest'])
    return df
```

```python
distance = get_cities_combination(ville)
distance['distance'] = distance[['origin','dest']].apply(lambda x: get_distance_between(*x), axis=1)
```

```python
distance = distance[distance.distance !=0]
distance.sort_values(by=['distance'],ascending= True)
```
