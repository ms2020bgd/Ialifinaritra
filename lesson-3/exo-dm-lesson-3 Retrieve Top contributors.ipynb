##  1. TOP contributors by crawling :

```python
import requests
from bs4 import BeautifulSoup

url = "https://gist.github.com/paulmillr/2657075"
```

```python
def get_soup_from_url(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    return soup
```

```python
soup = get_soup_from_url(url)
table=soup.find('table').find_all('tr')[1::1]

users = []

for items in table:
    data = items.find_all(['th','td'])
    name = data[1].a.text
    users.append(name)
    print(name)
```

## 2. Using Github API :

```python
import requests
from collections import defaultdict
from github import Github
import pandas as pd

g = Github("d7554474e4045658d4495d38fbff079713029844")
note = {}

for i in range(20) :
    stars = 0
    nb_repos = 0
    for repos in g.get_user(users[i]).get_repos():
        stars = stars + repos.stargazers_count
    mean_stars = stars / len(list(g.get_user(users[i]).get_repos()))
    note[users[i]] = mean_stars
    i += 1

    
```

```python
df = pd.DataFrame.from_dict(note, orient='index')
df.columns = ['means stars']
df.sort_values(by=['means stars'],ascending= False)
```
